```python
import requests # Sends HTTP requests to get webpage content 
from bs4 import BeautifulSoup as bs #Parses and extracts HTML content (like tags, text, links)
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_ministries_and_departments_of_the_Government_of_India'
page = requests.get(url) #  Sends a GET request to the given URL
page.status_code # Returns HTTP status code (200 = success)
```

```python
page.content #Returns the raw HTML of the page in bytes
```

```python
"""
soup = bs(page.content,'html.parser') #Converts HTML into a searchable object. 'html.parser' is the built-in parser
soup.prettify() # Formats the HTML nicely for easier reading.
"""
```

```python
soup = bs(page.text, "lxml")
```

```python
soup
```

```python
section_header = soup.find(id="Current_Ministries")
current_table = section_header.find_next("table")
```

```python
import io
df_current = pd.read_html(io.StringIO(str(current_table)))[0]
```

```python
df_current
```

```python
df_current.to_csv("web_scraping_minister_list_wikipedia.csv",index=False)
```
